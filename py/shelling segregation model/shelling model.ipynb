```python
import matplotlib.pyplot as plt
import random
import time
import numpy as np
import shelling_matrix as m

def create_data():  
    matrix = m.make_matrix(80, 6350, 0.5)
    plt.pcolor(matrix)
    plt.show()
    return matrix

def life_cycle(matrix, agents, hap, rounds = 1000):
    cycDone=0
    happiness = []
    happiness.append(hap)
    for i in range(rounds):
        if m.all_happy(agents):
            break
        matrix = m.cycle(agents, matrix, tolerance)
        agents = m.make_agent_list(matrix, tolerance)
        happiness.append(m.happiness_coef(agents))
        cycDone += 1
        if cycDone % 30 == 0:
            print("Result of ", cycDone, " cycles")
            plt.pcolor(matrix)
            plt.show()
    plt.pcolor(matrix)
    plt.show()
    xs = range(0, cycDone+1)
    plt.plot(xs, happiness)
    
    
        
        
tolerance = 0.5
matrix=create_data()
agents = m.make_agent_list(matrix, tolerance)
happiness = m.happiness_coef(agents)
life_cycle(matrix, agents, happiness, 20)
```
