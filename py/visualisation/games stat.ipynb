```python
import json
import matplotlib.pyplot as plt
from collections import Counter

with open('pract3/GAMES.json', encoding='utf8') as f:
    games = json.loads(f.read()) # Таблица решений задач

years = [i['Year'] for i in games]
cnt = list(Counter(years).items())
cnt.sort(key=lambda i : i[0])
x = [int(i[0]) for i in cnt if i[0] != 'не издана']
y = [i[1] for i in cnt if i[0] != 'не издана']
plt.figure()
plt.bar(x, y)
plt.title("Популярнейшие годы");
plt.grid(True)
```

```python
data = [[i['Year'], i['Genre']] for i in games if i['Year'] != 'не издана']
genres = set([i[1] for i in data])
dictPlots = {}
x = range(1981, 2009)
plt.figure(figsize=(10,10))
for g in genres:
    tmp = [0] * 28
    for i in data:
        if i[1] == g:
            tmp[int(i[0])-1981] += 1
    plt.plot(x, tmp, label=g)
plt.legend();
```
